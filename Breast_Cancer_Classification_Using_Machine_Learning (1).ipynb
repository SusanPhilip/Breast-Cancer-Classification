Importing Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# Loading data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data into a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
# Printing the first 5 rows of the data
data_frame.head()
```

```python
# Adding the target column to the data frame
data_frame['Diagnosis'] = breast_cancer_dataset.target
```

```python
# Print last five rows of the data
data_frame.tail()
```

```python
# Printing number of rows and columns in the dataset
data_frame.shape
```

```python
# Getting some information about the data
data_frame.info()
```

```python
# Checking for missing values
data_frame.isnull().sum()
```

```python
# Getting statistical measures about the data
data_frame.describe()
```

```python
# Checking the distribution of target variables
data_frame['Diagnosis'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
# Plotting a graph based on a few parameters
sns.pairplot(data_frame,hue = 'Diagnosis', palette= 'coolwarm', vars = ['mean radius', 'mean texture', 'mean perimeter','mean area','mean smoothness'])
```

```python
# visualize the count:
sns.countplot(data_frame['Diagnosis'], label = 'count')
```

```python
data_frame.groupby('Diagnosis').mean()
```

Seperating the features and target

```python
X = data_frame.drop(columns='Diagnosis', axis=1)
Y = data_frame['Diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression model using Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# Accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on Training data is ',training_data_accuracy)
```

```python
# Accuracy on Testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on Testing data is ',testing_data_accuracy)
```
